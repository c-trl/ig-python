##Working with IG's API: Collecting More with Pagination

In a request to IG's API, 3 parent dictionaries are returned: pagination, meta, and data.  Calling data returns a forced maximum of 30-32 instagram posts relevant to a search query (specified by the url fed into a requests command).  Calling pagination returns information about the 'page' of data returned in data.  Pagination also returns information about the next 'page', which can be called upon by referring to the url under 'next_url'.  Running the following:

returns 'https://api.instagram.com/v1/tags/newyorkcity/...'

Theoretically, a script can be created which accepts an initial url, and then calls that url's 'next url'.  This could ultimately be looped x times, returning 30*(x) rows of data.

```python
import pandas as pd
import requests
import json

client_id = '768fcf1f36c94eb08506bae0a9caffa3'
secret = '54efcbaed7f64673bc93b4e28ca9e8b2'
base_url = "https://api.instagram.com/v1"
query = "newyorkcity"

url = '{0}/tags/{1}/media/recent?client_id={2}&count=100'.format(base_url, query, client_id)

r = requests.get(url).json()
```

```python
def ig_comb(initial_url):
    from instagram.client import InstagramAPI
    import requests
    import urllib2
    import json
    import pandas as pd
    from pandas.io.json import json_normalize
        

    df = json_normalize(requests.get(initial_url).json()['data'])

    url2 = json_normalize(requests.get(initial_url).json()['pagination'])['next_url'][0]
    df2 = json_normalize(requests.get(url2).json()['data'])

    url3 = json_normalize(requests.get(url2).json()['pagination'])['next_url'][0]
    df3 = json_normalize(requests.get(url3).json()['data'])
    
    url4 = json_normalize(requests.get(url3).json()['pagination'])['next_url'][0]
    df4 = json_normalize(requests.get(url4).json()['data'])
    
    url5 = json_normalize(requests.get(url4).json()['pagination'])['next_url'][0]
    df5 = json_normalize(requests.get(url5).json()['data'])
    
    url6 = json_normalize(requests.get(url5).json()['pagination'])['next_url'][0]
    df6 = json_normalize(requests.get(url6).json()['data'])
    
    url7 = json_normalize(requests.get(url6).json()['pagination'])['next_url'][0]
    df7 = json_normalize(requests.get(url7).json()['data'])
    

        
        
    all_pages = [df2,df3,df4,df5,df6,df7]
    
    return df.append(all_pages).reset_index().drop('index',axis=1)
```
